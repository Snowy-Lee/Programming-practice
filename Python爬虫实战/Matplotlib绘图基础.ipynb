```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
x = np.linspace(0,100,1000)
```

```python
y = np.sin(x)
```

```python
plt.plot(x)
```

```python
#  xxyclwls()
plt.plot(x,y,c = 'red',lw = 1.2,ls = '--')#lw1.2--
plt.show()
```

```python
plt.plot(x,y,ls = '-',marker = 'o',markersize=1.2,c='red',\
        markeredgecolor='blue',markerfacecolor = 'black',label ='xy')
plt.show()
```

```python
#
plt.rcParams['font.sans-serif'] = ['SimHei']
#
plt.rcParams['axes.unicode_minus']=False
x = np.linspace(0,10,100)
y = np.sin(x)
plt.plot(x,y,ls=':',lw=2,label='y = sin(x)')#label
plt.legend(loc='center')#
plt.show()
```

```python
x = np.arange(0,1.1,0.01)
```

```python
y = x**2
```
