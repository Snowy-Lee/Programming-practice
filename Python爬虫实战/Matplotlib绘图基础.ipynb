```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
x = np.linspace(0,100,1000)
```

```python
y = np.sin(x)
```

```python
plt.plot(x)
```

```python
#绘制线图  x：x轴数据，y轴数据，c线条颜色，lw线条宽度，ls线条风格(实线、虚线、点画线、实点线)
plt.plot(x,y,c = 'red',lw = 1.2,ls = '--')#lw通常1.2，点画线--，实点线：
plt.show()
```

```python
plt.plot(x,y,ls = '-',marker = 'o',markersize=1.2,c='red',\
        markeredgecolor='blue',markerfacecolor = 'black',label ='x和y的关系')
plt.show()
```

```python
#解决中文显示乱码问题
plt.rcParams['font.sans-serif'] = ['SimHei']
#解决符号显示问题
plt.rcParams['axes.unicode_minus']=False
x = np.linspace(0,10,100)
y = np.sin(x)
plt.plot(x,y,ls=':',lw=2,label='y = sin(x)')#label是标签
plt.legend(loc='center')#图例
plt.show()
```

```python
x = np.arange(0,1.1,0.01)
```

```python
y = x**2
```

```python
#画布大小(8,9)、分辨率（越大越清晰）、边框颜色
plt.figure(figsize=(6.4,4.84),dpi = 100,facecolor='white')#black，blue
plt.title('这是一幅图')
plt.xlabel('x1')
plt.ylabel('y')
plt.xlim([0,1])#坐标轴范围
plt.ylim([0,1])
plt.xticks([0,0.2,0.4,0.6,0.8,1])#刻度
plt.yticks([0,0.2,0.4,0.6,0.8,1])
plt.plot(x,y,label='y = x^2')
plt.legend(loc='best')
plt.savefig('F:\jupyter file\Python爬虫实战\可视化.png')#pdf,
plt.show()
```
